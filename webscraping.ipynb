```python
import requests
from random import choice

url = 'http://httpbin.org/user-agent'

with open('user_agent.txt') as file:
    lines = file.read().split('\n')

for line in lines:
    user_agent = {'user-agent': choice(lines)}
    response = requests.get(url=url, headers=user_agent)
    print(response.text)
```

```python
from random import choice
import requests

url = 'http://httpbin.org/ip'

with open('proxy.txt') as file:
    proxy_file = file.read().split('\n')
    for _ in range(1000):
        try:
            ip = choice(proxy_file).strip()
            proxy = {
                'http': f'http://{ip}',
                'https': f'https://{ip}'
            }
            response = requests.get(url=url, proxies=proxy)
            print(response.json(), 'Success connection')
        except Exception as _ex:
            continue
```

```python
import requests
from bs4 import BeautifulSoup

# Отправляем GET-запрос к странице
url = "https://example.com"
response = requests.get(url)

# Если запрос успешен, извлекаем содержимое страницы
if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")

    # Ищем элементы с заданными классами
    price_box = soup.find("div", class_="price_box")
    price = price_box.find("p", class_="price").text

    # Выводим результат
    print(f"Цена: {price}")
else:
    print("Произошла ошибка при запросе")
```

```python
from bs4 import BeautifulSoup
import requests

url = 'https://parsinger.ru/html/index3_page_1.html'
response = requests.get(url=url)
response.encoding = 'utf-8'
soup = BeautifulSoup(response.text, 'lxml')

all_lists = []

# Получаем ссылки на все страницы
links = soup.find('div', 'pagen').find_all('a')
shema = 'https://parsinger.ru/html/'
urls = [f"{shema}{link['href']}" for link in links]

# Проходимся по каждой странице
for url in urls:
    response = requests.get(url)
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text, 'lxml')
    
    div = soup.find('div', 'item_card')
    names = div.find_all('a', class_='name_item')
    title_list = [name.text for name in names]
    all_lists.append(title_list)

print(all_lists)
```
