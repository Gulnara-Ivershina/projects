```python
import pandas as p
import numpy as np
df = pd.read_csv('horse_data.csv')
df.head()
```

```python
df_new = df.iloc[:, [0,1,3,4,5,6,10,22]]
```

```python
df_new.columns = ['surgery','age','rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome']
df_new.head()
df_new.info()
```

```python
df_new = df_new.replace('?', float('nan'))
df_new.head()
```

```python

df_new['rectal temperature'] = df_new['rectal temperature'].astype('float')
df_new['pulse'] = df_new['pulse'].astype('float')
df_new['respiratory rate'] = df_new['respiratory rate'].astype('float')
df_new.info()
```

```python
df_new.head(100)
```

```python
df_new.describe()
```

```python
mode = df_new.mode().iloc[0]
disp = round(df_new.var(numeric_only=True), 2)
print(mode, disp, sep='\n')
```

```python
iqr = df_new.quantile(0.75) - df_new.quantile(0.25)
```

```python
lower_bound = df_new.quantile(0.25) - 1.5*iqr
upper_bound = df_new.quantile(0.75) + 1.5*iqr
outliers = df_new[(df_new < lower_bound) | (df_new > upper_bound)]
outliers.head(10)
```

```python
###Во-первых, удаляю все строки, где есть больше трех значений  NAN, потери небольшие (около 20 строк из 300)
###'rectal_temperature' (60 пропусков) -  можно поставить среднюю температуру во всех пропусках, т.к. судя по значениям температура колеблется около 38 градусов;
###'respiratory rate' (58 пропусков) - частота дыхания, очень зависит от возраста лошади и от спокойного состояния, тут возникает большой разброс данных,
###поэтому использую моду.
###'age' - возраст лошади, тут нет пропусков, но в таблице какая-то неразбериха, должно быть 1 и 2, а возникает какая-то 9, вероятно это опечатка и должно быть 2 вместо 9
###'surgery' - наличие операции, всего 1 пропуск, т.к. лошадь старая, предположила, что была операция, поэтому заменила на константу.
###'pain' - боль, непонятно, как заполнять пропуски, заполнила модом.
```

```python
df_new = df_new[df_new.isnull().sum(axis=1) <= 2]
df_new.head(299)
```

```python
df_new['surgery'].fillna('1', inplace = True)
df_new['age'] = df_new['age'].replace(9, 2)
df_new['rectal temperature'].fillna(df_new['rectal temperature'].mode()[0], inplace = True)
df_new['pulse'].fillna(df_new['pulse'].mode()[0], inplace = True)
df_new['respiratory rate'].fillna(df_new['respiratory rate'].median(), inplace = True)
df_new['temperature of extremities'].fillna(df_new['temperature of extremities'].mode()[0], inplace = True)
df_new['pain'].fillna(df_new['pain'].mode()[0], inplace = True)
df_new['outcome'].fillna('4', inplace = True)
df_new.head(277)
```
