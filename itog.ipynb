```python
data = {
    'country_code': ['USA','FRA','DEU','GBR','CHN','VNM','RUS','KAZ','MNA','KGZ'],


}
```

```python
df_countries = pd.DataFrame(data = data)
df_countries.head(11)
```

```python
df_new = pd.read_sql(
    """
    SELECT
    name,population,country_code
    FROM
    city;

    ""","postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
df_new.head()
```

```python
# Объединяем два датафрейма для того, чтобы привязаться к странам из нашей гугл-таблицы и получить новую таблицу
```

```python
d1 = pd.DataFrame(df_countries)
d2 = pd.DataFrame(df_new)
int_df = pd.merge(d1,d2,how = 'inner',on = ['country_code'])
int_df.head()
```

```python
# Выводим таблицу с городами из нашей гугл-таблицы и населением этих городов
```

```python
df_city = pd.merge(df_csv,int_df,how='inner',left_on='City',right_on='name')
df_city.head(11)
```

```python
df_population = df_city[['country_code','population']]
```
