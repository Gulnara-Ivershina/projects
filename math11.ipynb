```python
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
#Уменьшение объекта в два раза
resize = np.array([[0.5, 0, 0], [0, 0.5, 0], [0, 0, 1]])
A_scaled = np.dot(A, resize)
A_scaled
```

```python
#Поворот на 130 градусов

rotate_angle = np.radians(130)
rotate_transform = np.array([[np.cos(rotate_angle), -np.sin(rotate_angle), 0],[np.sin(rotate_angle), np.cos(rotate_angle), 0], [0, 0, 1]])
A_rotated = np.dot(A, rotate_transform)
```

```python
# Отражение объекта относительно прямой y=x
reflection_transform = np.array([[0, 1, 0],
                                 [1, 0, 0],
                                 [0, 0, 1]])

A_reflected = np.dot(A, reflection_transform)
```

```python
# Построение графика для каждого преобразования

fig, ax = plt.subplots(1, 4, figsize=(16, 4))

# Исходный объект
ax[0].plot(A[:,0], A[:,1])
ax[0].set_xlim([-200, 200])
ax[0].set_ylim([-200, 200])
ax[0].set_title('Исходный объект')

# Уменьшенный объект
ax[1].plot(A_scaled[:,0], A_scaled[:,1])
ax[1].set_xlim([-200, 200])
ax[1].set_ylim([-200, 200])
ax[1].set_title('Уменьшенный в два раза')

# Повернутый объект
ax[2].plot(A_rotated[:,0], A_rotated[:,1])
ax[2].set_xlim([-200, 200])
ax[2].set_ylim([-200, 200])
ax[2].set_title('Повернутый на 130 градусов')

# Отраженный объект
ax[3].plot(A_reflected[:,0], A_reflected[:,1])
ax[3].set_xlim([-200, 200])
ax[3].set_ylim([-200, 200])
ax[3].set_title('Отраженный относительно y=x')

plt.show()
```

```python
e_values, e_vectors = np.linalg.eig(m)

print("Собственные значения:")
print(e_values)
print("\nСобственные векторы:")
print(e_vectors)
```

```python
K = np.array([[-0.85065081, -0.52573111],
              [0.52573111, -0.85065081]])

M = np.array([[-0.23606798, 0],
              [0, 4.23606798]])

K_inv = np.linalg.inv(K)

m = np.dot(np.dot(K, M), K_inv)

print("Матрица m с использованием спектрального разложения:")
print(m)

```
