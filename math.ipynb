```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#Задание 1 Уровень 0
x = sym.Symbol('x')

y = sym.Symbol('y')

# Функция f(x, y) = 2x^2y^3 + 1/x + y^2x + 7
f1 = 2*x**2*y**3 + 1/x + y**2*x + 7

# Вычисление частной производной по x
df1_dx = sym.diff(f1, x)
print("Частная производная по x:", df1_dx)

# Вычисление частной производной по y
df1_dy = sym.diff(f1, y)
print("Частная производная по y:", df1_dy)

# Функция f(x, y) = x^2y - sin(xy) + cos(x^2) + 6y
f2 = x**2*y - sym.sin(x*y) + sym.cos(x**2) + 6*y

# Вычисление частной производной по x
df2_dx = sym.diff(f2, x)
print("Частная производная по x:", df2_dx)

# Вычисление частной производной по y
df2_dy = sym.diff(f2, y)
print("Частная производная по y:", df2_dy)
```

```python
#Уровень 1 Задание 2
# Вычислить градиент функции f(x,y)=2x^2y^3 + 1/x + y^2x + 7 в точке (1,2)
gradient_x = df1_dx.subs([(x, 1), (y, 2)]).evalf()
gradient_y = df1_dy.subs([(x, 1), (y, 2)]).evalf()

print("Градиент функции в точке (1, 2):", gradient_x, gradient_y)
```

```python
# Преобразование символьного выражения f в функцию
f_func = sym.lambdify((x, y), f1)

# Создание сетки точек для отображения функции
X, Y = np.meshgrid(np.linspace(-10, 10, 100), np.linspace(-10, 10, 100))
Z = f_func(X, Y)
```

```python
# Визуализация функции и градиента
plt.figure()
plt.contourf(X, Y, Z, levels=20, cmap='RdYlBu')
plt.colorbar(label='f(x, y)')
plt.quiver(1, 2, float(gradient_x), float(gradient_y), scale=20, color='white')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Visualization')
plt.grid()
plt.show()
```

```python
#Задание 3
def f(x1, x2):
    return x2**2 - x1**2 * x2 + 2 * x1 * x2
```

```python
x1 = np.linspace(-100, 100, 1000)
x2 = np.linspace(-100, 100, 1000)
X1, X2 = np.meshgrid(x1, x2)
```

```python
Z = f(X1, X2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='viridis')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1, x2)')
ax.set_title('Трехмерный график функции')
plt.show()
```
