```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
# Определяем символьные переменные
x = symbols('x')

# Определяем функцию первой производной
f1 = 10*sin(2*x+1)**4*cos(2*x+1)

# Создаем функцию, возвращающую численные значения первой производной
f1_numerical = lambdify(x, f1, 'numpy')

# Создаем массив значений x
x_vals = np.linspace(-5, 5, 100)

# Создаем массив значений первой производной
y_vals = f1_numerical(x_vals)

# Строим график первой производной
plt.plot(x_vals, y_vals)
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("График первой производной f(x) = 10*sin**4(2x+1)*cos(2x+1)")
plt.grid(True)
plt.show()
```

```python
# Определяем функцию второй производной
f2 = 80*sin(2*x+1)**3*cos(2*x+1)**2 - 20*sin(2*x+1)**5

# Создаем функцию, возвращающую численные значения второй производной
f2_numerical = lambdify(x, f2, 'numpy')

# Создаем массив значений второй производной
y2_vals = f2_numerical(x_vals)

# Строим график второй производной
plt.plot(x_vals, y2_vals)
plt.xlabel('x')
plt.ylabel("f''(x)")
plt.title("График второй производной f(x) = 80*sin**3(2x+1)*cos**2(2x+1) - 20*sin**5(2x+1)")
plt.grid(True)
plt.show()
```
